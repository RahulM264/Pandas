# Sorting values

```python
#Sorts names in alphabetical order if ascending = True else in descending order if ascending = false
df.sort_values('Name',ascending=False)
```

```python
#Soring Attacks in decreasing order
df.sort_values('Attack',ascending=False)
```

```python
#Now, Soring both Attack and Defense highest in only bug type pokemon
bug_df = df[df['Type 1']=='Bug']
bug_df.sort_values(['Attack','Defense'],ascending=[1,1])
```

# Changing Values and Adding Columns


```python
#First we located the numeric content and indexed the numeric content required and added to get total
df['Total'] = df.iloc[:,4:10].sum(axis=1)
df
```

```python
#Here we created a list of columns and added by indexing them to obtain data frame with columns interchanged
cols = list(df.columns)
df = df[cols[0:4]+[cols[-1]]+cols[4:12]]
df
```

# Filtering data

```python
#here we can see the indexes are not correct so we can change the index by taking this to a new data set and resetting index
new_df = df[(df['Type 1']=='Electric') & (df['Legendary']==True) & (df['Attack']>10)]
new_df = new_df.reset_index()
```

```python
#We used the reset index funciton to reset the index of this new data frame
new_df
```

```python
#Filtering the names which contain mega
df.loc[df['Name'].str.contains('Mega')]
```

```python
#Fire or Grass
df.loc[(df['Type 1']=='Fire') | (df['Type 1']== 'Grass')]
```

```python
#Easier way is to use the Re function
import re
df[df['Type 1'].str.contains('Fire|Grass',regex=True)]
```

```python
#Changed the name of Fire type to Flamer
df.loc[df['Type 1']=='Flamer', 'Type 1'] = 'Fire'
df

```

```python
#renamed the name of column to Dex Number
df.rename(columns={'#':'Dex Number'},inplace=True)
df
```

# Aggregate Statistics using Groupby Function


```python
#We use groupby function to get the statistical values of various groups.Here we see all the values of all groups of pokemon
df.groupby(['Type 1']).mean()
```

```python
#Now we sort this data based on a factor of HP and Attack
df.groupby(['Type 1']).mean().sort_values(['Attack','HP'],ascending=[1,1])
```
