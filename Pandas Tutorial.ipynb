# Head and Tail Function

Now that we have succesfully read the data, the first step would be to take a look at the data to understand the contents of it and get an idea of the basic information present in the data like Name of Columns, No. of Rows etc


This can be achieved by using the Head and Tail functions in Pandas

```python
draft1.head()
```

As we can see, the head() function returns the top 5 rows present in the dataset. This is a useful function because we can quickly get an idea of the contents inside the dataset

Similarly, tail() function returns the last 5 rows by default

```python
draft1.tail(10)
```

If the value is not specified, head() and tail() returns the first and the last 5 rows respecitvely but if the values are specified, they return the suitable number of rows accordingly as shown above

There are many parameters which are available inside read_csv() so lets look at a few of them right now

1. header = 'infer'
2. names = None
3. index_col = None



## header = 'infer' and names = None

By default, when we read any dataset, the parameter header is set to be 'infer'. Infer stands for inferring the header names from the dataset itself.

If no names are passed as column names in the read_csv() function, header becomes equal to 'infer' and Pandas automatically pass in the first row elements as column names by default. In this case, names are set to None

If names are passed explicitly, header becomes equal to none and those names are used as Column Names.

Let us see this with the help of examples

```python
draft2 = pd.read_csv('pokemon_data.csv')
```

Here, the names of the columns are not explicitly mentioned, so header is equal to infer and column names are the first row values of the dataset

```python
colnames = list(draft1.columns)
colnames.append('New Name')


draft3 = pd.read_csv('pokemon_data.csv',names=colnames)
```

```python
draft3
```

Here, the column names are explicitly mentioned and passed in the read_csv() function so this time header becomes equal to None and these names are used as column names. 

Notice that the usual column names occupied the first row this time

## index_col

As the name implies, index_col refers to the index column of the dataframe. By default, index_col is set to None and indexing usually starts from 0 in this case till the end of number of rows.

There are two ways in which we can index the dataframe in Pandas

1. Single Indexing
2. Multi Indexing

```python
draft1.head()
```

```python
ex0 = pd.read_csv('pokemon_data.csv',index_col='#')

ex0.head()
```

Notice the differences in draft1 and ex0 dataframes.

In draft1, we did not specify the index column so the indexing started from 0 and continues onwards till the end.

But in ex0, since we explicitly mentioned index_col = '#' column, the '#' column is set as index so in this case, indexing started from 1.

This is an example of Single Indexing

```python
ex1 = pd.read_csv('pokemon_data.csv',index_col = ['#','Legendary'])

ex1.head(20)
```

ex2 dataframe is an example of Multi Indexing wherein the dataframe has two indexes namely '#' and 'Legendary'

These are the various types of indexing we can use in Pandas

# isnull() function

The isnull() function is one of the most used functions while working with Pandas Dataframes. Often the data we obtain from the real-world is messy and it has a lot of NaN or Null Values that comes along with it. 

The isnull() function returns the Dataframe consisting of Trues and Falses based on if the data present in that location is null or not.

Let us see the example

```python
draft1.isnull()
```

Just by looking at this dataframe, we can see a clear shortcoming. Although this does return the true or falses, since the data is huge, we cannot clearly see how many columns have null values

To overcome this problem, we use sum() function which returns the number of null values in each column of the dataframe

```python
draft1.isnull().sum()
```

As we can see, Type 2 Column in this dataframe has 386 null values.

## Reading from URL

Let us read the data from a url which contains the data in the form of tables and not as raw data. Inorder to read the data from such sources, we use read_html() function to return the table of data


```python
url = 'https://www.basketball-reference.com/leagues/NBA_2015_totals.html'

exa1 = pd.read_html(url)

exa1
```

```python
type(exa1)
```

Even though the code worked properly, we can see that the data is not present in the form of a dataframe, instead it is in the form of a list. 

This is because of the fact that, when we use read_html() function on a url, it returns the data which are present in tables. These tables can be one or many depending on the website.

But, since it is difficult to display all the tables at once, this function appends all the dataframes in a list, essentially wrapping them.

So inorder to access individual dataframes, we need to index the list

```python
dfa = exa1[0]
dfa
```

Now we can see that this data is a dataframe

# Selecting one or multiple columns

In Pandas, we can select one or multiple columns by passing their names in a list within the dataframe.

Lets look at how we do it

```python
draft1['Name']
```

```python
type(draft1['Name'])
```

```python
draft1[['Name']]
```

```python
type(draft1[['Name']])
```

As we can notice from the past 4 cells, if we mention the name of the column within a single [], pandas interpret this data as Pandas Series and not Pandas Dataframe. Inorder for us to make this into a dataframe, we have to pass it as a list with the [].

The same is the case for multiple columns selecting. If we pass multiple column names without a list, we will encounter an error since a Pandas Series can only have one column

```python
draft1[['Name','Type 1']]
```
